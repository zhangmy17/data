```python
class HashTable:
    """
    哈希函数实现哈希表
    """
    def __init__(self, size=10):
        self.size = size
        self.table = [[] for _ in range(self.size)]  # 初始化桶列表

    def _hash_function(self, key):
        """
        计算键的哈希值
        """
        return hash(key) % self.size  # 使用Python内置hash函数取模

    def insert(self, key, value):
        """
        插入键值对
        """
        index = self._hash_function(key)
        # 检查是否已存在该键，若存在则更新值
        for pair in self.table[index]:
            if pair[0] == key:
                pair[1] = value
                return
        # 不存在则添加新键值对
        self.table[index].append([key, value])

    def search(self, key):
        """
        查找键对应的值
        """
        index = self._hash_function(key)
        for pair in self.table[index]:
            if pair[0] == key:
                return pair[1]
        return None

    def delete(self, key):
        """
        删除键值对
        """
        index = self._hash_function(key)
        for i, pair in enumerate(self.table[index]):
            if pair[0] == key:
                del self.table[index][i]
                return True
        return False
```

```python
if __name__ == "__main__":
    ht = HashTable()
    ht.insert("name", "Alice")
    ht.insert("age", 25)
    print(ht.search("name"))
    print(ht.search("age"))
    ht.delete("age")
    print(ht.search("age"))   #
```
