```python
# ====================== 最小生成树 MST：Prim & Kruskal 实现 ======================
# 图数据：顶点 A,B,C,D,E,F，所有无向带权边
edges = [
    ("A", "B", 2),
    ("A", "D", 3),
    ("B", "C", 4),
    ("B", "E", 1),
    ("D", "E", 6),
    ("E", "F", 2),
    ("C", "F", 5)
]
vertices = ["A", "B", "C", "D", "E", "F"]
n = len(vertices)

# ---------------------- Kruskal 算法（并查集实现） ----------------------
class UnionFind:
    def __init__(self, vertex_list):
        self.parent = {v: v for v in vertex_list}

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):
        x_root = self.find(x)
        y_root = self.find(y)
        if x_root != y_root:
            self.parent[y_root] = x_root
            return True
        return False

def kruskal(edge_list, vertex_list):
    # 1. 边按权重升序排序
    sorted_edges = sorted(edge_list, key=lambda x: x[2])
    uf = UnionFind(vertex_list)
    mst = []
    total_weight = 0
    print("===== Kruskal 算法执行步骤 =====")
    for idx, (u, v, w) in enumerate(sorted_edges):
        print(f"第{idx+1}轮，检查边 ({u},{v})，权重={w}")
        if uf.union(u, v):
            mst.append((u, v, w))
            total_weight += w
            print(f"选中该边，当前MST集合：{mst}，累计权重：{total_weight}")
            if len(mst) == len(vertex_list) - 1:
                break
        else:
            print(f"舍弃，会形成环")
    return mst, total_weight

# ---------------------- Prim 算法（邻接表，起点A） ----------------------
def build_adj(edge_list):
    adj = {v: [] for v in vertices}
    for u, v, w in edge_list:
        adj[u].append((v, w))
        adj[v].append((u, w))
    return adj

def prim(start, edge_list, vertex_list):
    adj = build_adj(edge_list)
    visited = set()
    mst = []
    total_weight = 0
    visited.add(start)
    print("\n===== Prim 算法执行步骤（起点A） =====")
    step = 1
    while len(visited) < len(vertex_list):
        min_edge = None
        min_w = float("inf")
        # 遍历所有已访问点的邻边，找最小跨集边
        for u in visited:
            for v, w in adj[u]:
                if v not in visited and w < min_w:
                    min_w = w
                    min_edge = (u, v, w)
        u_sel, v_sel, w_sel = min_edge
        mst.append(min_edge)
        total_weight += w_sel
        visited.add(v_sel)
        print(f"第{step}步：选中跨集边 ({u_sel},{v_sel})，权重={w_sel}")
        print(f"当前连通集S：{sorted(list(visited))}，MST边：{mst}，累计权重：{total_weight}")
        step += 1
    return mst, total_weight

# ====================== 运行计算 ======================
# Kruskal 结果
kruskal_mst, kruskal_sum = kruskal(edges, vertices)
print(f"\nKruskal 最小生成树总权重：{kruskal_sum}")
print(f"MST边集合：{kruskal_mst}")

# Prim 结果（起点A）
prim_mst, prim_sum = prim("A", edges, vertices)
print(f"\nPrim 最小生成树总权重：{prim_sum}")
print(f"MST边集合：{prim_mst}")
```

结果：

1. 两种算法得到的最小生成树完全一致

2. MST 边：('B', 'E', 1), ('A', 'B', 2), ('E', 'F', 2), ('A', 'D', 3), ('B', 'C', 4)

3. MST 总权重：1+2+2+3+4 = 12
