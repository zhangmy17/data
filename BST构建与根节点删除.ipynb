1.构建BST

插入 50：空树，50 作为根节点。

插入 30：30<50，成为 50 的左孩子。

插入 74：74>50，成为 50 的右孩子。

插入 20：20<50→到左子 30，20<30，成为 30 的左孩子。

插入 40：40<50→到左子 30，40>30，成为 30 的右孩子。

插入 60：60>50→到右子 74，60<74，成为 74 的左孩子。

插入 80：80>50→到右子 74，80>74，成为 74 的右孩子。

          50
        /    \
      30      74
     /  \    /  \
    20  40  60  80

2.中序前驱替换

找到 50 的中序前驱：左子树最右下节点40。

把 40 的值放到根节点位置，再删除原来的 40 节点，40 是叶子，直接删。

          40
        /    \
      30      74
     /       /  \
    20      60  80

3.中序后继替换

找到 50 的中序后继：右子树最左下节点60。

把 60 的值放到根节点位置，再删除原来的 60 节点，60 是叶子，直接删。

          60
        /    \
      30      74
     /  \       \
    20  40       80

4.混用

最好不混用。

混用会让树的高度波动无规律，破坏 BST 删除操作的一致性，后续查找、删除的时间复杂度难以稳定控制。工程实现里通常固定只用一种策略（要么统一用前驱、要么统一用后继），方便代码维护和树的平衡性维持。极端频繁混用会让树退化成链表形态，查找效率从O(log n)退化到O(n)。

5.怎么删最好

追求树更平衡：优先选左右子树规模更大的那一侧，能让树高度更小，后续操作更快。

代码实现统一：全程固定一种策略，避免维护混乱。

频繁删除场景：交替在左右选前驱或后继可以延缓树退化，但要设计规范的交替规则，不能无规律混用。
