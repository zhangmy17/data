1.插入30

仅根节点 30，BF(30)=0

      30(BF=0)

 无失衡

2.插入20

20<30，作为 30 左孩子

        30(BF=1)
       /
    20(BF=0)

BF(30)=1-0=1，|BF|≤1，无失衡

3.插入10

10<20，作为 20 左孩子

          30(BF=2)  ←失衡，LL型，旋转轴30
         /
        20(BF=1)
       /
    10(BF=0)

BF(30)=2>1，LL 型失衡，旋转轴为 30，对 30 做右旋

右旋后：

      20(BF=0)
     /  \
    10  30(BF=0)

BF(20)=0，BF(10)=BF(30)=0，恢复平衡。

4.插入25

25 介于 20 和 30 之间，作为 30 左孩子

       20(BF=1)
     /    \
    10   30(BF=1)
         /
        25(BF=0)

 BF(20)=1-1=0，BF(30)=1-0=1，全部|BF|≤1，无失衡。

5.插入40

40>30，作为 30 右孩子

       20(BF=1)
     /    \
    10   30(BF=0)
         /  \
        25  40(BF=0)

BF(30)=1-1=0，BF(20)=1-1=0，无失衡

6.插入35

35 介于 30 和 40 之间，作为 40 左孩子

       20(BF=1)
     /    \
    10   30(BF=-1)
         /  \
        25  40(BF=1)
            /
           35(BF=0)

从下往上算平衡因子：BF(40)=1，BF(30)=1-2=-1，BF(20)=1-2=-1，|BF|均≤1，无失衡

7.插入50

50>40，作为 40 右孩子

        20(BF=1)
       /        \
     10      30(BF=-2)  ←失衡，RR型，旋转轴30
             /    \
            25   40(BF=-1)
                 /   \
                35  50(BF=0)

BF(30)=1-3=-2，RR 型失衡，旋转轴为 30，对 30 做左旋

        20(BF=1)
       /        \
     10      40(BF=0)
             /    \
           30(BF=0) 50
           /  \
         25  35

BF(10)=0,BF(25)=0,BF(35)=0,BF(50)=0,BF(30)=1-1=0,BF(40)=2-2=0,BF(20)=1-2=-1，全部|BF|≤1，平衡恢复

8.中序遍历
中序遍历规则：左→根→右

遍历顺序：10 → 20 → 25 → 30 → 35 → 40 → 50

满足BST左小右大的性质
