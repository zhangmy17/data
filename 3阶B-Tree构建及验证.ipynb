3.性质验证
关键字数量约束：

   根节点：1≤2≤2

   非根叶子节点：每个节点关键字数为1或2，满足1≤k≤2

子节点数量约束：k个关键字对应k+1个子节点：根2个关键字对应3个子节点；叶子无关键字对应0个子节点

有序性：同一节点内关键字升序，左子<父关键字<右子，整体有序

叶子层级：所有叶子都在最底层，层数一致

最大子节点数：根最多3个子节点，符合m=3上限
