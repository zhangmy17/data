```python
from typing import Any
from collections.abc import Sequence

class OrderedArrayInsert:
    """
    有序数组插入
    """

    def __init__(self, iterable: Sequence[Any]) :
        """
        :param iterable: 有序数组初始化
        """
        self.iterable = list(iterable)


    def insert(self, element: int) -> None:
        """
        插入元素
        :param element: 待插入元素
        """

        insert_index: int = len(self.iterable)


        for index, value in enumerate(self.iterable):
            if element < value:
                insert_index = index
                break

        self.iterable.append("")
        self.len = len(self.iterable)
        for i in range(self.len - 1, insert_index , -1):
            self.iterable[i] = self.iterable[i - 1]

        self.iterable[insert_index] = element
```

```python
    base_array: list[int] = [1,4,21,59]

    ordered_array_insert = OrderedArrayInsert(base_array)
    ordered_array_insert.insert(25)
    ordered_array_insert.insert(69)
    ordered_array_insert.insert(62)
```
