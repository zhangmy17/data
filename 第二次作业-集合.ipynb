```python
from collections.abc import Sequence
from typing import Any

class MySet:
    """
    集合的查询、插入、删除
    插入、删除前检查是否存在
    """
    def __init__(self,iterable:Sequence[Any]):
        """

        :param iterable: 初始化集合，实际类型为列表
        """
        self.iterable = list(iterable)

    def insert_last(self,element:Any) -> None:
        """

        :param element: 要插入的元素
        :return: 插入是否成功
        """

        for index,item in enumerate(self.iterable):
            if item == element:
                return False
        self.iterable.append(element)
        return True

    def delete_set(self,element:Any) -> None:
        """

        :param element: 要删除的元素
        :return: 删除是否成功
        """

        for index,item in enumerate(self.iterable):
            if item == element:
                self.iterable.remove(element)
                return True

        return False


    def find_set(self,element:Any) -> None:
        """

        :param element: 要查找的元素
        :return: 是否找到
        """

        for index,item in enumerate(self.iterable):
            if item == element:
                return True

        return False
```

```python
fruits = ['apple','banana','orange','mango','pineapple']
my_set = MySet(fruits)
print(my_set.insert_last('banana'))
print(my_set.iterable)
print(my_set.find_set('banana'))
print(my_set.iterable)
print(my_set.delete_set('banana'))
print(my_set.iterable)
print(my_set.find_set('banana'))
print(my_set.iterable)
```
