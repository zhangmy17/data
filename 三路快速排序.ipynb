```python
import random

def quick_sort_3way(arr, low, high):
    if low >= high:
        return

    # 随机选轴，避免最坏情况
    pivot_index = random.randint(low, high)
    arr[pivot_index], arr[low] = arr[low], arr[pivot_index]
    pivot = arr[low]

    lt = low      # arr[low+1 .. lt] < pivot
    i = low + 1   # 当前扫描元素
    gt = high     # arr[gt .. high] > pivot

    while i <= gt:
        if arr[i] < pivot:
            arr[lt], arr[i] = arr[i], arr[lt]
            lt += 1
            i += 1
        elif arr[i] > pivot:
            arr[i], arr[gt] = arr[gt], arr[i]
            gt -= 1
        else:
            i += 1

    # 递归处理小于区和大于区（等于区已就位）
    quick_sort_3way(arr, low, lt - 1)
    quick_sort_3way(arr, gt + 1, high)

# 包装调用
def three_way_quick_sort(arr):
    quick_sort_3way(arr, 0, len(arr) - 1)
```

```python
test_arr = [2, 0, 2, 1, 1, 0, 2, 2, 1, 0, 1, 2, 0, 0, 0, 1, 1]
three_way_quick_sort(test_arr)
print(test_arr)
```
