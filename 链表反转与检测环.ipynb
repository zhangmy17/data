1.链表反转

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverse(head: ListNode) -> ListNode:
    pre = None
    cur = head
    while cur:
        # 先保存下一个节点
        next_tmp = cur.next
        # 反转当前节点的指针
        cur.next = pre
        # pre和cur同步后移
        pre = cur
        cur = next_tmp
    return pre  # pre最终成为新头结点

def reverse_recur(head: ListNode) -> ListNode:
    # 递归终止：空节点/单个节点直接返回
    if not head or not head.next:
        return head
    # 递归反转后半段链表
    new_head = reverse_recur(head.next)
    # 翻转后一个节点的指向
    head.next.next = head
    head.next = None
    return new_head

# 构建 a->b->c 链表
a = ListNode('a')
b = ListNode('b')
c = ListNode('c')
a.next = b
b.next = c

new_head = reverse(a)
# 遍历输出 c b a
p = new_head
while p:
    print(p.val, end=' ')
    p = p.next
```

2.检测环

```python
def has_cycle(head: ListNode) -> bool:
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False

def has_cycle_set(head: ListNode) -> bool:
    visited = set()
    cur = head
    while cur:
        if cur in visited:
            return True
        visited.add(cur)
        cur = cur.next
    return False

# 构造带环链表：a->b->c->b
a = ListNode('a')
b = ListNode('b')
c = ListNode('c')
a.next = b
b.next = c
c.next = b

print(has_cycle(a))  # 输出 True

# 构造无环链表
x = ListNode('x')
y = ListNode('y')
x.next = y
print(has_cycle(x))  # 输出 False
```
